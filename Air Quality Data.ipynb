```python
#Split the data into Training set and Test set
from sklearn.model_selection import train_test_split 
#importing the scikitlearn package and importing the class train_test_split class
```

```python
X=df[0]
y=df[1]
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size=0.2, random_state = 1)
```

```python
# 0 here means column name OZONE
X_train
```

```python
#1 means column name Solar.R
y_train
```

```python
#rotating the matrix on x so reshaping on (-1,1)
#flipping values is necesary because for linear regression we need to perform matrix multiplication with train and test values accordingly...
X_train=X_train.values.reshape((-1,1))
```

```python
#importing the linea model class from scikit learn package and perfroming linear regression
from sklearn import linear_model as lm
m=lm.LinearRegression()
results=m.fit(X_train,y_train)
```

```python
#to get the accuracy of the model
accuracy=m.score(X_train,y_train)
print('accuracy of the model is:',accuracy)
```

```python
#Print coeffecients (printing the slope and the intercepts)
print('intercept:',m.intercept_)
print('slope:',m.coef_)
```

# The slope value 0.879 means that predicted ozone increases by 0.879

```python
X_test=X_test.values.reshape((-1,1))
```

```python
#predictions from the model
predictions=m.predict(X_test)
print('predicted Ozone values:',predictions,sep='\n')
```

```python
#plotting a scatterplot b\w test data and predicted data
import matplotlib.pyplot as plt
plt.scatter(y_test,predictions)
```

```python
#Evaluating the Model
from sklearn.metrics import mean_squared_error, r2_score
X_train=X_train.reshape(-1,1)
y_train_prediction=m.predict(X_train)
X_test=X_test.reshape(-1,1)
y_test_prediction=m.predict(X_test)
```

```python
# Model evaluation for Training Set
import numpy as np
rmse_training=(np.sqrt(mean_squared_error(y_train,y_train_prediction)))
r2_training=r2_score(y_train,y_train_prediction)

print('The model performance for training set')
print('----------------------------------------')
print('RMSE is{}'.format(rmse_training))
print("R2 score is{}".format(r2_training))
```

```python
# Model evaluation for Testing Set
import numpy as np
rmse_testing=(np.sqrt(mean_squared_error(y_test,y_test_prediction)))
r2_testing=r2_score(y_test,y_test_prediction)

print('The model performance for testing set')
print('----------------------------------------')
print('Root Mean Sqaured Error: {}'.format(rmse_testing))
print("R2 score is {}".format(r2_testing))
```
